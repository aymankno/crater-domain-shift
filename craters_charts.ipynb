```python
import os
os.environ['CURL_CA_BUNDLE'] = ''
os.environ['REQUESTS_CA_BUNDLE'] = ''
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
mars_crater_img = Image.open("samples/sample_mars_crater.jpg")
moon_crater_img = Image.open("samples/sample_moon_crater.png")
```

```python
mars_crater_img
```

```python
moon_crater_img
```

```python
mars_img_small = mars_crater_img.resize((224, 224))
mars_array = np.array(mars_img_small)

moon_img_small = moon_crater_img.resize((224, 224))
moon_array = np.array(moon_img_small)
```

```python
# MARS CRATER HISTO
mars_gray = mars_img_small.convert("L")
mars_gray_array = np.array(mars_gray)

plt.hist(mars_gray_array.flatten(), bins = 50, color="gray")
plt.show()
```

```python
# MOON CRATER HISTO
moon_gray = moon_img_small.convert("L")
moon_gray_array = np.array(moon_gray)

plt.hist(moon_gray_array.flatten(), bins=50, color="gray")
plt.show()
```
